```python
# Copyright (c) OpenMMLab. All rights reserved.
import asyncio
import glob

from mmdet.apis import (async_inference_detector, inference_detector,
                        init_detector, show_result_pyplot)


class Args:
    img = "rummy-5/test/116-nastolnaya-igra-rummikub-dorozhnyj-v-metallicheskoj-korobke-rummikub-travel_jpg.rf.430f143a4912d9d6625fc6653335fb96.jpg"
    config = "config.py"
    checkpoint = "work_dirs/config/latest.pth"
    score_thr = 0.9
    device = 'cuda:0'
    palette = "coco"

def main(args):
    # build the model from a config file and a checkpoint file
    model = init_detector(args.config, args.checkpoint, device=args.device)
    # test a single image
    result = inference_detector(model, args.img)
    # show the results
    show_result_pyplot(
        model,
        args.img,
        result,
        palette=args.palette,
        score_thr=args.score_thr)

if __name__ == '__main__':
    args = Args()
    for img in glob.glob("rummy-5/test/*.jpg"):
      args.img = img
      main(args)
```
